```python
# Install Ollama Python package if not done yet
# !pip install ollama

from ollama import Client
import csv

# Initialize Ollama client
client = Client()

# Choose your model
model = "translategemma:12b"

# Single test sentence
sentence = "Hello, how are you?"

# Build translation prompt
prompt = f"""
You are a professional English (en) to Arabic (ar) translator.
Your goal is to accurately convey the meaning and nuances of the original English text
while adhering to Arabic grammar, vocabulary, and cultural sensitivities.
Produce only the Arabic translation, without any additional explanations or commentary.
Please translate the following English text into Arabic:


{sentence}
"""

# Run model
response = client.generate(model=model, prompt=prompt, options={'num_predict': 500})

# Extract output
translation = response.get("response", "").strip()

# Display result
print("English:", sentence)
print("TranslateGemma:12B:", translation)

# Optional: save to CSV
with open("translation_test.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["English", "TranslateGemma_12B"])
    writer.writeheader()
    writer.writerow({"English": sentence, "TranslateGemma_12B": translation})

print("Translation saved to translation_test.csv")
```
